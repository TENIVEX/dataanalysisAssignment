```python
import matplotlib.pyplot as plt
import numpy as np

with_music = [304,257,174,214,69,317,387,47,157,0,332,308,317,286,236,299,206,278,188,43,0,0,0,0,0]
without_music = [292,270,47,288,524,292,364,316,287,75,282,149,274,319,213,3,324,2,128,219,94,164,0,0,0]

xpos = np.arange(len(with_music))

plt.hist(with_music, bins=10, edgecolor='black', color='yellow', alpha=0.6, label='with music')
plt.hist(without_music, bins=10, edgecolor='black', color='blue', alpha=0.6, label='without music')
plt.title('Histogram of the growth with both music and without music')
plt.xlabel('Growth')
plt.ylabel('Frequency')
plt.legend(loc='best')
plt.show()
```
